```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
season_code = "25_26"
current_dir = Path.cwd() 
path_folder = current_dir.parent.parent.parent / "csv" / f"csv{season_code}" / "scouting"
leagues_list_names = ["Scouting"]
```

```python
def line_index(df):
    weights = {
        "Goals Against": 10,
        "Saves": 10,
        "Clean Sheets": 5,
        "Defensive Actions Outside Penalty Area": 5,
        "Crosses Stopped": 5,
        "Penalties Winner": 5,
        "Efficiency": 60,
        "% Saves": 25,
        "% Crosses Stopped": 5
    }
    total_weights = sum(weights.values())
    return df[list(weights.keys())].multiply(pd.Series(weights), axis=1).sum(axis=1) / total_weights
```

```python
def passes_index(df):
    weights = {
        "Launched Passes Completed": 40,
        "Passes Attempted (GK)": 20,
        "Through Balls": 20,
        "% Long Passes": 20,
    }
    total_weights = sum(weights.values())
    return df[list(weights.keys())].multiply(pd.Series(weights), axis=1).sum(axis=1) / total_weights
```

```python
def create_indices_df(df_centiles):
    static_cols = ['Player', 'Nation', 'General Position', 'Team', 'Age', 'Born', 'Matches Played', 'Starts', 'Minutes Played']
    
    df_indices = df_centiles[static_cols].copy()
    
    df_indices["Line Index"] = line_index(df_centiles).round().astype(int)
    df_indices["Passes Index"] = passes_index(df_centiles).round().astype(int)

    return df_indices
```

```python
for i, league_names in enumerate(leagues_list_names):
    path = path_folder / f"{leagues_list_names[i]}_centiles_gk.csv"
    path_end_indices = path_folder / f"{leagues_list_names[i]}_metrics_gk.csv"
    df_centiles = pd.read_csv(path)
    df_indices = create_indices_df(df_centiles)
    df_indices.to_csv(path_end_indices, index=False)
```
